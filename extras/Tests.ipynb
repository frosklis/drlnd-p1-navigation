```python
import sys
if '../' not in sys.path:
    sys.path.append('../')
from bananas import logger
from bananas.agents import DQN_Agent, Double_DQN_Agent
from bananas.problem import moving_average, moving_average_cross, loggable, decay
from unityagents import UnityEnvironment
from bananas.environment import UnityAdapter
from bananas.model import QNetwork, DuelingQNetwork
import mlflow
mlflow.set_tracking_uri('file:../mlruns')
banana_environment = UnityAdapter(
    UnityEnvironment(
        file_name="/Users/claudio/Dropbox/udacity-drl/code/deep-reinforcement-learning/p1_navigation/Banana.app",
        no_graphics=True,
    ),
    name='bananas')
```

```python
from sklearn.model_selection import ParameterGrid
import random
param_grid = {
    'agent': [DQN_Agent, Double_DQN_Agent],
    'decay_factor': [0.995, 0.9995],
    'update_every': [4, 7, 10],
    'model': [QNetwork, DuelingQNetwork]}
params = list(ParameterGrid(param_grid))
random.shuffle(params)
```

```python
for p in params:
    print(p)
    agent = p['agent'](
        banana_environment,
        is_finished_function=[loggable(moving_average_cross, 100, 200), loggable(moving_average, 100, 13)],
        is_solved=loggable(moving_average, 100, 13),
        max_train_episodes=2000,
        epsilon_update=loggable(decay, p['decay_factor'], min_value=0.01),
        update_every=p['update_every'],
        partial_network=p['model'],
        seed=0,
    )
    agent.train()
```

```python
agent = Double_DQN_Agent(
    banana_environment,
    is_finished_function=[loggable(moving_average_cross, 100, 200), loggable(moving_average, 100, 13)],
    is_solved=loggable(moving_average, 100, 13),
    max_train_episodes=2000,
    epsilon_update=loggable(decay, 0.8, min_value=0.01),
    seed=0)
agent.train()
```

```python
agent = DQN_Agent(
    banana_environment,
    is_finished_function=[loggable(moving_average_cross, 100, 200), loggable(moving_average, 100, 13)],
    is_solved=loggable(moving_average, 100, 13),
    max_train_episodes=2000,
    epsilon_update=loggable(decay, 0.8, min_value=0.01),
    seed=0)
agent.train()
```
